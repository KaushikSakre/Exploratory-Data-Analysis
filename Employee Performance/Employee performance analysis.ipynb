```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.compose import ColumnTransformer

transformer = ColumnTransformer(transformers=[
    ('tnf2',OrdinalEncoder(categories=[['Travel_Rarely', 'Travel_Frequently', 'Non-Travel']]),['BusinessTravel']),
    ('tnf3',OneHotEncoder(sparse_output=False,drop='first'),['Gender','MaritalStatus','Department','OverTime','JobRole','EducationField'])
],remainder='passthrough')
```

```python
X_train=transformer.fit_transform(X_train)
```

```python
X_test=transformer.transform(X_test)
```

```python
le = LabelEncoder()
```

```python
le.fit(y_train)
y_train = le.transform(y_train)
y_test = le.transform(y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rgclf = RandomForestClassifier(n_estimators=100,
                              random_state=3,
                              max_samples=0.5,
                              max_features=0.75,
                              max_depth=15)
```

```python
rgclf.fit(X_train,y_train)
```

```python
y_pred = rgclf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
